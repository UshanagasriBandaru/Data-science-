```python
import seaborn as sns
import pandas as pd
```

```python


sales = pd.read_excel('/content/sales.xlsx')
```

```python
sales.head()
```

```python
sns.relplot(x='Sales', y='Profit', data=sales)
```

```python
sns.relplot(x='Sales', y='Profit', data=sales,hue='Order Priority')#hue helps in colour coding
```

```python
sns.relplot(x='Sales', y='Profit', data=sales,hue='Order Priority',style='Order Priority')
```

```python
sns.relplot(x='Sales', y='Profit', data=sales,size='Discount')
```

```python
sns.relplot(x='Sales', y='Profit', data=sales,size='Discount',sizes=(25,200))
```

```python
sns.relplot(x='Order Date',y='Sales',data=sales,kind='line')
```

```python
sales['year'] = sales['Order Date'].dt.year#to extract year value
sales.year.unique()
```

```python
sns.relplot(x='year',y='Sales',data=sales,kind='line')
```

```python
sns.relplot(x='year',y='Sales',data=sales,kind='line',ci=None)#ci=confidence interval
```

```python
sns.relplot(x='year',y='Sales',data=sales,kind='line',ci='sd',estimator=None)#for accurate values
```

```python
sns.relplot(x='year',y='Sales',data=sales,kind='line',ci='sd',estimator=sum)
```

```python
sns.relplot(x='year',y='Sales',data=sales,kind='line',hue='Ship Mode')
```

```python
sns.relplot(x='year',y='Sales',data=sales,kind='line',hue='Ship Mode',style='Product Category')
```

```python
sns.relplot(x='year',y='Sales',data=sales,kind='line',hue='Ship Mode',col='Product Category',ci=None)
```

```python
sns.relplot(x='year',y='Sales',data=sales,kind='line',hue='Ship Mode',col='Region',ci=None)
```

```python
sns.relplot(x='year',y='Sales',data=sales,kind='line',hue='Ship Mode',col='Region',ci=None,col_wrap=4)
```
