```python
mean_value = df['CLOSE'].mean()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7,7))
```

```python
plt.hist(df['CLOSE'], bins=30, alpha=0.7, label='Data Distribution')
plt.axvline(mean_value, color='r', linestyle='dashed', linewidth=2, label=f'Mean: {mean_value:.2f}')
```

```python
plt.xlabel('OPEN')
plt.ylabel('CLOSE')
plt.title('Histogram with Mean Line')
plt.legend()
plt.show()
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("/content/FINAL_FROM_DF.csv")
```

```python
data_corr=dataset.select_dtypes(include=['number']).corr()
data_corr
```

```python
dataset.select_dtypes(include=['number']).cov()
```

```python
sns.heatmap(data_corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
